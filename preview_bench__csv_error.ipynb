```python
import pandas as pd
import numpy as np 
import cv2
import os
from pathlib import Path
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
workingdir = "/home/hbpopos/dealer-apps/apps/dealer-image-detector-card/src"
data_back = pd.read_csv(os.path.join(workingdir,"bench_card_back_errors.csv"))
for fname,detect,actual in zip(data_back["filename"],data_back["detection"],data_back["actual"]):
    print(Path(fname).stem.split()[0] + os.path.splitext(fname)[1])
    print(f"detect: {detect}, actual: {actual}")
    img = cv2.cvtColor(cv2.imread(os.path.join(workingdir,fname)), cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
from penis_utils import detecting_shit as ds
from penis_utils import finding_shit as fs
import os
from pathlib import Path
import shutil 
```

# tell if image is blurry , for manual labeling

```python
IMAGE_PATH = "C:\\card_localization\\card_raw"
all_images = fs.FilesFinder(IMAGE_PATH).by_extension(".jpg")
if not os.path.exists(os.path.join(IMAGE_PATH , "blurry")):
    os.makedirs(os.path.join(IMAGE_PATH , "blurry"))
if not os.path.exists(os.path.join(IMAGE_PATH , "not_blurry")):
    os.makedirs(os.path.join(IMAGE_PATH , "not_blurry"))
for img in all_images:
    check_blur = ds.ClassifyShit(img).tell_if_blurry(250)
    fname = Path(img).stem + os.path.splitext(img)[1]
    blur_path = os.path.join(IMAGE_PATH, "blurry")
    nb_path = os.path.join(IMAGE_PATH, "not_blurry")
    if check_blur == True:
        shutil.move(img, os.path.join(nb_path, fname))
    else:
        shutil.move(img, os.path.join(blur_path, fname))
    # print(check_blur, os.path.join(blur_path, fname))
    
```

# class balance check
run the cell below to check class balance of your labled items. 

```python
from counting import CountClasses
def get_class_balance(path):    
    penis = CountClasses(path=path)
    penis.count_classes()
    print("classes: " , len(penis.count_class))
    penis.draw_plot()
get_class_balance("C:\\baccarat_finetune\\testc")
```
